```python
df['Date']=df['Day']+' '+df['Month']
df.drop(columns=['Day','Month'], axis=1,inplace=True)
df['Series'] = df['Series'].replace('\\n',' ', regex=True)
```

# 6.Scrape the details of top 100 songs on billiboard.com.

```python
url=('https://www.billboard.com/charts/hot-100')
driver=webdriver.Chrome(r"D://Chromedriver.exe")
driver.get(url)
```

```python
#Creating empty lists
Song=[]
Artist=[]
Last_Week_Rank=[]
Peak_Rank=[]
Weeks_on_board=[]
```

```python
#Scraping Song Name
ntag=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
for i in ntag:
    if i.text is None :
        Song.append("--") 
    else:
        Song.append(i.text.replace("/"," "))
        

#Scraping Artist
atag=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
for i in atag:
    if i.text is None :
        Artist.append("--") 
    else:
        Artist.append(i.text)
        
last_rank_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for i in last_rank_tag:
    if i.text is None :
        Last_Week_Rank.append("--") 
    else:
        Last_Week_Rank.append(i.text)
        
        
#scraping peak rank
peak_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for i in peak_tag:
    if i.text is None :
        Peak_Rank.append("--") 
    else:
        Peak_Rank.append(i.text)
        

#Scraping Weeks on board
wob=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for i in wob:
    if i.text is None :
        Weeks_on_board.append("--") 
    else:
        Weeks_on_board.append(i.text)
```

```python
# creating the dataframe from the scraped data
import pandas as pd
TOP100SONGS=pd.DataFrame({})
TOP100SONGS['Song_name']=Song
TOP100SONGS['Artist_name']=Artist
TOP100SONGS['Last_Week_Rank']=Last_Week_Rank
TOP100SONGS['Peak_Rank']=Peak_Rank
TOP100SONGS['Weeks_on_board']= Weeks_on_board
TOP100SONGS
```
